# Librerias

```python
from skimage import io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from skimage.feature import local_binary_pattern
from skimage.transform import rotate
```

```python
img = io.imread("tema12_act.png" , as_grey=True)
io.imshow(img)
np.shape(img)

img1 = np.random.rand(1098,1100)
img2 = np.random.rand(1098,1100)
img3 = np.random.rand(1098,1100)

for y in range(1098):
    for x in range(1100):
        img1[y][x]=img[y][x]

for y in range(1098):
    for x in range(1100, 2200):
      
        img2[y][x-1100]=img[y][x]

for y in range(1098):
    for x in range(2200, 3300 ):
   
        img3[y][x-2200]=img[y][x]


plt.figure(1)
plt.subplot(1,3,1)
plt.imshow(img1 , cmap="gray")
plt.subplot(1,3,2)
plt.imshow(img2, cmap="gray")
plt.subplot(1,3,3)
plt.imshow(img3, cmap="gray")
        
```

# LBP

utilizar LBP uniformes sobre estas texturas

```python
METHOD = 'uniform'
P = 24
R = 3  
lbp_text1 = local_binary_pattern(img1, P,R, METHOD)
lbp_text2 = local_binary_pattern(img2, P,R, METHOD)
lbp_text3 = local_binary_pattern(img3, P,R, METHOD)

plt.figure(1)
plt.subplot(1,3,1)
plt.imshow(lbp_text1 , cmap="gray")
plt.subplot(1,3,2)
plt.imshow(lbp_text2, cmap="gray")
plt.subplot(1,3,3)
plt.imshow(lbp_text3, cmap="gray")
```

```python
np.unique(lbp_text1.ravel())
```

# histograma 

Se concluye que la figura 1 y 3 son similares

```python
plt.figure(1)
plt.subplot(1,3,1)
plt.hist(lbp_text1.ravel(), normed=True, bins=P+2)
plt.subplot(1,3,2)
plt.hist(lbp_text2.ravel(), normed=True, bins=P+2)
plt.subplot(1,3,3)
plt.hist(lbp_text3.ravel(), normed=True, bins=P+2)
```

```python
textura1_rotada = rotate(img1,angle=90,resize=False)
io.imshow(textura1_rotada)
```

```python
lbp_textrotada1 = local_binary_pattern(textura1_rotada, P,R, METHOD)
plt.imshow(textura1_rotada, cmap="gray")
```

# Propiedad de la Invarianza

el resultado del mismo es independiente de la orientaci√≥n de la textura o imagen de entrada. Se pide, por tanto, demostrar de forma pr√°ctica esta propiedad de invarianza.

```python
plt.figure(1)
plt.subplot(3,1,1)
plt.title("ROTADA")
plt.hist(lbp_textrotada1.ravel(), normed=True, bins=P+2)

plt.subplot(3,1,3)
plt.title("ORIGINAL")
plt.hist(lbp_text1.ravel(), normed=True, bins=P+2)
```
