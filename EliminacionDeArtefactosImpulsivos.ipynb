### Trabajo 3 - Eliminaci√≥n de artefactos impulsivos en una imagen
#### Iniciamos importando las bibliotecas necesarias para la actividad

```python
from skimage import filters
from skimage import io
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
```

#### Cargamos la imagen a procesar y la mostramos

```python
img = data.imread("tema7_actividada.jpg")
io.imshow(img)
```

#### Se crea una funci√≥n que modela la adici√≥n, de forma aleatoria, de artefactos impulsivos. El resultado tendr√° que apreciarse en la visualizaci√≥n de la imagen, que deber√° estar afectada por ruido de tipo ¬´sal y pimienta¬ª.

```python
for i in range(7000):
    img[np.random.randint(0, 500)][np.random.randint(0, 500)] = 255
    img[np.random.randint(0, 500)][np.random.randint(0, 500)] = 0
io.imshow(img)
```

#### A partir de la imagen ruidosa, buscaremos eliminar los artefactos impulsivos para el posterior tratamiento de la imagen, para ello vamos a utilizar uno de los filtros disponibles en la biblioteca skimage

```python
img_sin_ruido = filters.median(img)
io.imshow(img_sin_ruido)
```

#### A partir de la imagen obtenida en la etapa anterior, se busca identificar las siluetas de las estructuras en ella presentes y como en el paso anterior haremos uso del filtro sobel que encuentra la magnitud del borde usando la transformada de Sobel

```python
silueta = filters.sobel(img_sin_ruido)
io.imshow(silueta)
```
